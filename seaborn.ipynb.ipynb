2.Create a scatter plot of total_bill vs tip using the 'tips' dataset.

```python
#Scatter Plot
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title("Total_Bill vs Tips")
plt.show()
```

3. Plot a line graph for total_bill vs tip using Seaborn.

4. Draw a bar plot to show the average tip for each day in the 'tips' dataset.

```python

#barplot
sns.barplot(data=df, x='day', y='tip', estimator='mean')

# Add labels and title
plt.title('Average Tip by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip')

# Show plot
plt.show()
```

5. Use a count plot to display how many records are available for each day.

```python

#countplot
sns.countplot(data=df, x='day')

# Add labels and title
plt.title('Number of Records per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Count')

# Show plot
plt.show()
```

6. Make a box plot to show total_bill distribution across different days.

15. Create a heatmap of the correlation matrix from the 'penguins' dataset using numeric_only=True

```python

# Load the 'penguins' dataset
penguins = sns.load_dataset('penguins')

# Compute the correlation matrix for numeric columns only
corr_matrix = penguins.corr(numeric_only=True)

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f', square=True)

# Add a title
plt.title('Correlation Heatmap - Penguins Dataset')

# Show plot
plt.show()
```

20. Convert all categorical columns in the 'tips' dataset to numeric and create a full correlation heatmap.

```python
# Convert all categorical columns to numeric using one-hot encoding
tips_numeric = pd.get_dummies(df, drop_first=True)

# Compute the full correlation matrix
corr_matrix = tips_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)

# Add title
plt.title('Correlation Heatmap - Tips Dataset (All Numeric)')

# Show plot
plt.show()
```
